# 数据分析示例

本 Notebook 展示了如何使用本项目的数据分析工具。

## 1. 导入库

```python
import sys
sys.path.append('../app')

import pandas as pd
import numpy as np
from data_loader import DataLoader
from analyzer import DataAnalyzer
from visualizer import DataVisualizer

print('✓ 所有库导入成功')
```

## 2. 加载数据

```python
# 创建示例数据
data = {
    '日期': pd.date_range(start='2024-01-01', periods=100, freq='D'),
    '销售额': np.random.randn(100).cumsum() + 1000,
    '访问量': np.random.randint(100, 1000, 100),
    '转化率': np.random.uniform(0.01, 0.1, 100),
    '产品类别': np.random.choice(['A', 'B', 'C', 'D'], 100)
}

df = pd.DataFrame(data)
df.head()
```

## 3. 数据分析

```python
# 创建分析器
analyzer = DataAnalyzer(df)

# 查看统计信息
stats = analyzer.get_statistics()
stats
```

```python
# 查看基本信息
info = analyzer.get_info()
print(f"数据集形状: {info['shape']}")
print(f"列名: {info['columns']}")
print(f"内存使用: {info['memory_usage'] / 1024:.2f} KB")
```

```python
# 计算相关性
corr = analyzer.get_correlation()
corr
```

## 4. 数据可视化

```python
# 创建可视化器
visualizer = DataVisualizer(df)

# 绘制销售额分布
visualizer.plot_histogram('销售额', bins=30, title='销售额分布')
visualizer.show()
```

```python
# 绘制散点图
visualizer.plot_scatter('访问量', '销售额', title='访问量 vs 销售额')
visualizer.show()
```

```python
# 绘制时间序列
visualizer.plot_time_series('日期', '销售额', title='销售额时间序列')
visualizer.show()
```

```python
# 绘制相关性热力图
visualizer.plot_heatmap(title='特征相关性')
visualizer.show()
```

## 5. 高级分析

```python
# 按产品类别分组分析
group_stats = analyzer.group_by_and_aggregate('产品类别', '销售额', 'mean')
group_stats
```

```python
# 查找异常值
outliers = analyzer.find_outliers('销售额', method='iqr')
print(f"发现 {len(outliers)} 个异常值")
outliers.head()
```
